```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("./marketing_campaign1.csv")
df.head()
```

```python
df.info()
```

```python
# null values in income
df["Income"] = df["Income"].fillna(df["Income"].mean())
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"],format='mixed').dt.year
```

```python
categorical_cols = ["Education", "Marital_Status"]

preprocessor = ColumnTransformer(
    transformers=[
        ( OneHotEncoder(), categorical_cols)
    ],
    remainder="drop"
)

encoded_features = preprocessor.fit_transform(df)
encoded_df = pd.DataFrame(encoded_features, columns=preprocessor.get_feature_names_out())
```

```python
# Select numerical columns for clustering
numerical_cols = df.select_dtypes(include=["number"])

# Scale numerical features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[numerical_cols])
scaled_df = pd.DataFrame(scaled_features, columns=numerical_cols)
```
